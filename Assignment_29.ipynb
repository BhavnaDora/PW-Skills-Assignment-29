after installing bokeh we import necessary libraries-

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()           # this function will help us to display bokeh plots inside jupyter notebook
from bokeh.plotting import figure, output_file, show
```

Then we take any x and y data and plot them using these steps-

```python
x=[1,2,3,4]
y=[2,3,4,1]
p= figure(title= "bokeh line plot")
p.line(x,y)
output_file("line.html")           # html file to display output
show(p)                            # to see the plot
```

In bokeh, Glyphs are visual shapes that are used to represent the data points in a plot. They define the basic geometry of the plot, such as the shape, size, and color of the markers used to represent the data.

```python
# adding circle glyph to the plot-
from bokeh.plotting import figure, output_file, show
bokeh.io.output_notebook

p= figure(title= "Circle glyph")
p.circle(x= [1,2,3,4], y=[2,4,6,8], size= 20, fill_color="red")
p.circle_cross(x=[1,2,3,4], y=[3,5,7,1], size= 20, fill_color="blue", fill_alpha= 0.3)

output_file("circle.html")
show(p)
```

We can customize bokeh plot by-
We can add title to the plot using the figure function's title parameter. p.figure(title= "bokeh plot")
We can add axes labels as- p.xais.axis_label= "my x axis" and similarly for y asis- p.yaxis.axis_label= "my y axis"
We can use the legend parameter in the bokeh plot. For ex) we are plotting a line graph- p.line(x= [], y=[], legend_label="these are data points")
We can also apply 5 in-built themes to the bokeh plot using the function- curdoc().theme= "dark_minimal" - this will give a dark background to our plot.

The Bokeh server is a component of Bokeh that allows you to build interactive web applications that are connected to Python code running on a server.
Bokeh creates high-level Python models, such as plots, ranges, axes, and glyphs, and then converts these objects to JSON to pass them to its client library, BokehJS. The primary purpose of the Bokeh server is to synchronize data between the underlying Python environment and the BokehJS library running in the browser.
first we build a simple application code let's say a program to plot sine wave.
This application code is served to the client browser. The code renders a sine wave and also has sliders to control frequency and amplitude, so when in the browser side, user tries to change the slider, the input hits the specific function of our application code i.e. new values are provided to backend code(application code) via the bokeh server. Then the plot with updated values is displayed in real time.

Ways to embed standalone Bokeh documents and Bokeh applications into web pages.

Standalone documents:

These documents dont require a Bokeh server to work. They may have many tools and interactions such as custom JavaScript callbacks but are otherwise nothing but HTML, CSS, and JavaScript. These documents can be embedded into other HTML pages as one large document or as a set of sub-components with individual templating.

Bokeh applications:

These applications require a Bokeh server to work. Having a Bokeh server lets you connect events and tools to real-time Python callbacks that execute on the server.
